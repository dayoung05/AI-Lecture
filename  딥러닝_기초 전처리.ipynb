## Scaling
### Min Max Scaling 
- 최대값 1, 최소값 0
- train/test split 전 pandas에서 직접 적용 가능
- train/test split 후 Sklearn 라이브러리 사용 가능

### Staadardization
- 표준편차 1, 평균 0
- 대개는 Min Max Scaling 많이 적용
- 이상치가 많은 경우 Standardization 적용

### Min Max Scaling 수식 적용

```python
df.columns
```

```python
for col in df.columns[:-1]:    # output은 수식에 적용되면 안됨
    df[col] = (df[col] - df[col].min())/(df[col].max()- df[col].min())
```

```python
df.head()
```

```python
df.describe()  # 최대값 1, 최소값 0으로 적용됨
```

### Standardizarion 수식 적용 

```python
df = pd.read_csv("4-2 heart.csv", index_col = 0)
```

```python
for col in df.columns[:-1]:
    df[col] = (df[col] - df[col].mean())/ df[col].std()
```

```python
df.describe()
```

## One-Hot Encoding
- Category Data를 그대로 문자 -> 숫자로만 바꿔서 넣으면 ML은 카테고리 데이터가 아닌 Value값으로 인식, ML학습이 정상적으로 이루어지지 않을 수 있음
- 따라서, Category Data는 One-Hot Encoding 처리 필수

### One Hot encoding 적용 예시

```python
titanic.head()
# embark_town: 카테고리 데이터 임으로 one hot encoding 적용
```

```python
titanic = pd.get_dummies(titanic, columns = ['embark_town'], prefix = 'embark_town')
# prefix 옵션을 설정하면 더미의 접두어를 변경할 수 있습니다
```

```python
titanic
```

## Train/Test Split
**Train/Test Split 개념**  
- 전체 Data를 학습용 train, 최종 성능 평가를 위한 test로 분리  
- Valid Data  
    - 모델의 학습 성능 중간 평가, 
    - 모델 Hypar Parameter를 Tuning 목적으로 생성
    
**Keras에서 Train/Test Split**  
- Keras의 Model fitting 과정에서, Validation_split이라는 Parameter 통해, 학습 과정 중 전체 Data 중 일부를 Valid Data로 쪼개서 Valid Data로 중간 Score를 측정하는 기능 가지고 있어, 따로 Valid Set을 만들지 않아도 된다

### Train/Test Split 실습
